```python
df_movies = df[df['type'] == 'Movie']
df_movies['release_year'].value_counts()
df_movies['rating'].value_counts()
df_movies.describe()
```

```python
good_2018_movies = df_movies[(df_movies['release_year'] == 2018)
                            & (df_movies['rating'] == 'PG-13')]
good_2018_movies.head(20)

sorted_good_2018_movies = good_2018_movies.sort_values(by='duration', ascending=False)
sorted_good_2018_movies.head(20)
#
```

```python
df_shows = df[df['type'] == 'TV Show']

df_shows.sort_values(by='duration', ascending=False).head(20)
```

```python
country_counts = df['country'].value_counts().head(20)


plt.figure(figsize=(13,6))

country_counts.plot(kind='bar')

plt.title('Countries of Production Distribution')
plt.xlabel('Countries')
plt.ylabel('Frequency')
plt.xticks(rotation = 90, ha='right')
plt.show
```

```python
release_year = df['release_year']

plt.figure(figsize=(13,6))
plt.hist(release_year, bins=5, edgecolor='black')

plt.title('Release Year Distribution')
plt.xlabel('Release Year')
plt.ylabel('frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
movies_duration = df_movies['duration'].sort_values(ascending=False)

plt.figure(figsize=(12,6))
plt.hist(movies_duration.head(2500), bins=10, edgecolor='black')
plt.title('Movies Duration Distribution')
plt.xlabel('Durations')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.show()
```

```python
genre_count = df['listed_in'].value_counts()

plt.figure(figsize=(12,6))
genre_count.head(20).plot(kind='bar')
plt.title('Top 20 Genres')
plt.xlabel('Genres')
plt.ylabel('Frequency')
plt.xticks(rotation=90, ha='right')
plt.show()
```
